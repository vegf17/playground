```python
import pennylane as qml
from pennylane import numpy as np
```

```python
dev = qml.device("default.qubit")

def W(params):
    (_, n_qubits) = np.shape(params)
    qml.BasicEntanglerLayers(weights=params, wires=range(n_qubits), rotation=qml.RY)

def S(g, x, num_wires):
    for i in range(num_wires):
        qml.exp(g(i),-1.0j * x)

@qml.qnode(dev)
def quantum_model(param_set, g, x):
    (n_train, _, n_qubits) = np.shape(param_set)
    for i in range(n_train-1):
        W(param_set[i])
        S(g,x,n_qubits)

    W(param_set[n_train-1])
    return qml.probs(wires=0)
        
```

```python
params = [[[1.0472, 0.7854, 3.1416, 0.3927],
           [1.0472, 0.7854, 3.1416, 0.5236]],
          [[1.0472, 0.7854, 1.5708, 0.3927],
           [0.7854, 0.7854, 1.5708, 0.7854]]]
gate = qml.X
gate_param = 0.7854

print(quantum_model(params, gate, gate_param))
print(qml.draw(quantum_model)(params, gate, gate_param))
```
