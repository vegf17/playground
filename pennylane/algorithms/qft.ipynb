# Quantum Fourier Transform

Quantum Fourier Transform (QFT) is:
* a quantum algorithm that maps a quantum state from the computational basis to the Fourier (frequency) basis
* mainly used as a subroutine inside other quantum algorithms such as:
    * Shor's algorithm (factoring, discrete logarithms)
    * Period-finding problems
    * Hidden subgroup problems 
    * Phase estimation

The QFT, rather than acting on classical data, acts on the amplitude structure of a quantum state.

The key role of the QFT is to transform a quantum state that encodes hidden periodicity into a form where that periodicity becomes visible as peaks in the measurement outcomes. 
In this way, periodic patterns in the amplitudes of a quantum state are converted into information that can be extracted by measurement.
Unlike the classical Fourier transform, the QFT does not output the full frequency spectrum directly; instead, it allows us to sample from it probabilistically.

The input to the QFT is a quantum state expressed in the computational basis, and the output is the same state expressed in the Fourier basis. 
A measurement performed after the QFT produces a single sample according to the resulting probability distribution, so the algorithm must typically be repeated multiple times to extract reliable information.

QFT has the following properties:
* Unitary operation
* Fourier-structure properties
    * convolution in the computation basis corresponds to pointwise multiplication in the Fourier basis
    * a cyclic shift of the input amplitudes results in a phase change in the Fourier basis, leaving the measurement probabilities unchanged
* Periodicity 
    * when a periodic function is encoded into the amplitudes of a quantum state, applying the QFT produces peaks at regularly spaced frequencies

Let us now dive a little bit into the maths behind the QFT.

The Quantum Fourier Transform, as said, is a unitary operation.
Hence, we can describe it as a matrix or as a function acting in the quantum states.

### Matrix formulation of the QFT

Let us begin by describing it as a matrix (as in the classical setting, the QFT can be seen as a basis change between the quantum states and the Fourier basis):
\begin{align*}
    U_{QFT} = \dfrac{1}{\sqrt{N}}
    \begin{pmatrix}
        1 & 1 & 1 & \dots & 1 \\
        1 & \omega & \omega^{2} & \dots & \omega^{(N-1)} \\
        1 & \omega^{2} & \omega^{4} & \dots & \omega^{2(N-1)} \\
        \vdots & \vdots & \vdots & \ddots & \vdots \\
        1 & \omega^{(N-1)} & \omega^{2(N-1)} & \dots & \omega^{(N-1)(N-1)}
    \end{pmatrix}
\end{align*}

where:
* $N = 2^{n}$ and $n$ is the number of qubits
* $\omega = e^{i \frac{2 \pi}{N}}$

### Function formulation of the QFT

Before presenting the function formulation of the QFT, it is important to note that a quantum state, in the computational-basis, can be interpreted as an integer. 
Recall that quantum states, in the computational basis, are strings formed by 0's and 1's, i.e. they are elements of $\{0,1\}^{n}$.
In other words, they are binary strings. 
An integer $x$ can be represented as a binary string as follows:
\begin{align*}
    x &= x_1 x_2 \dots x_{n-1} x_n \\ 
    &= x_1 \times 2^{n-1} + x_2 \times 2^{n-2} + \dots + x_{n-1} \times 2^{1} + x_{n} \times 2^{0} \\
    &= \sum_{k=1}^{n} x_{k} \times 2^{n-k}
\end{align*}

For instance, consider $n=2$.
We then have the following correspondences:
\begin{align*}
    \ket{0} &\Leftrightarrow \ket{00} \\
    \ket{1} &\Leftrightarrow \ket{01} \\ 
    \ket{2} &\Leftrightarrow \ket{10} \\
    \ket{3} &\Leftrightarrow \ket{11}
\end{align*}

It is important to recall this resemblance between the binary representation of a number and its integer representation, because the formulation of the QFT as a function uses both forms, as we shall see next.

Consider we have at our disposal $n$ qubits.
The action of $U_{QFT}$ on a $n$-qubit state is given by:
\begin{align*}
    U_{QFT} \ket{x} = \frac{1}{\sqrt{N}} \sum_{y=0}^{N-1} \omega^{x \times y} \ket{y}
\end{align*}
where $\ket{x}$ and $\ket{y}$ are quantum states in the binary notation while $x \times y$ is integer multiplication (i.e. $x$ and $y$ are in the integer notation)

Note that the QFT does not act on the bits of $x$ individually, but on the global integer value encoded by the computational-basis state $\ket{x}$.
In other words, the QFT acts on the integer representation of $x$ instead of acting on the bits that compose of $x$ individually.

## Example: QFT with one qubit

### Matrix formulation

With only one qubit the matrix of the QTF is as follows (note that $\omega = e^{i \pi} = -1$):
\begin{align*}
    U_{QFT} = \frac{1}{\sqrt{2}}
    \begin{pmatrix}
        1 & 1 \\
        1 & -1
    \end{pmatrix}
\end{align*}

Note that this matrix is the Hadamard gate.

A generic one-qubit state is $\begin{pmatrix} \alpha \\ \beta \end{pmatrix}$, where $\alpha, \beta \in \mathcal{C}$ and $|\alpha|^2 + |\beta|^2 = 1$.

Applying $U_{QFT}$ to $\psi$ gives the following:
\begin{align*}
    U_{QFT} \ket{\psi} 
    &= \frac{1}{\sqrt{2}}
    \begin{pmatrix}
        1 & 1 \\
        1 & -1
    \end{pmatrix} 
    \begin{pmatrix}
        \alpha \\
        \beta
    \end{pmatrix}
    = \frac{1}{\sqrt{2}}
    \begin{pmatrix}
        \alpha + \beta \\
        \alpha - \beta
    \end{pmatrix}
\end{align*}

Note that after applying the QFT, the amplitudes of the quantum states are now interacting with each other.

### Function formulation

With one qubit, the generic state is $\ket{\psi} = \alpha \ket{0} + \beta \ket{1}$, where $\alpha, \beta \in \mathcal{C}$ and $|\alpha|^2 + |\beta|^2 = 1$.

Applying $U_{QFT}$ to $\ket{\psi}$ gives us:
\begin{align*}
    U_{QFT} \ket{\psi} 
    &=
    U_{QFT} (\alpha \ket{0} + \beta \ket{1})\\
    &=
    \alpha U_{QFT} \ket{0} + \beta U_{QFT} \ket{1}\\
    &=
    \frac{1}{\sqrt{2}} \left( \alpha \sum_{y=0}^{1} \omega^{0 \times y} \ket{y} + \beta \sum_{y=0}^{1} \omega^{1 \times y} \ket{y} \right) \\
    &=
    \frac{1}{\sqrt{2}} \left( \alpha \left( \omega^{0} \ket{0} + \omega^{0} \ket{1} \right) + \beta \left( \omega^{0} \ket{0} + \omega^{1} \ket{1} \right) \right) \\
    &=
    \frac{1}{\sqrt{2}} \left( \alpha \left( \ket{0} + \ket{1} \right) + \beta \left( \ket{0} - \ket{1} \right) \right) \\
    &=
    \frac{\alpha + \beta}{\sqrt{2}} \ket{0} + \frac{\alpha - \beta}{\sqrt{2}} \ket{1}
\end{align*}


## From a mathematical expression to a quantum circuit

For this part, we need to introduce the concept of *fractional binary* of a number.
Intuitively, this concept allows to convert a decimal fraction to binary.
Given a number $$\frac{x_l}{2} + \frac{x_{l+1}}{2^{2}} + \dots + \frac{x_m}{2^{m-l+1}}$$ we can represent it as $$0.x_{l} x_{l+1} \dots x_{m}$$

For instance, the number $0.75= \frac{1}{2} + \frac{1}{2^2}$ is represented as $0.11$ in the fractional binary notation.

Now, we shall proceed to demonstrate how to go from a mathematical expression to a quantum circuit.
For that, we use the function formulation of the QFT:
\begin{align*}
    U_{QFT} \ket{x} 
    &=
    \frac{1}{\sqrt{N}} \sum_{y=0}^{N-1} \omega^{x \times y} \ket{y} \\
    &=
    \frac{1}{\sqrt{N}} \sum_{y=0}^{N-1} e^{i \frac{2 \pi}{N} x \times y} \ket{y} \\
    &\stackrel{(1)}{=}
    \frac{1}{\sqrt{N}} \sum_{y=0}^{N-1} e^{i \frac{2 \pi}{N} x \times \sum_{k=1}^{n} y_{k} 2^{n-k}} \ket{y} \\
    &=
    \frac{1}{\sqrt{N}} \sum_{y=0}^{N-1} \prod_{k=1}^{n} e^{i \frac{2\pi}{2^n} x \times y_k \frac{2^n}{2^k}} \ket{y} \\
    &=
    \frac{1}{\sqrt{N}} \sum_{y=0}^{N-1} \prod_{k=1}^{n} e^{i 2\pi x \times \frac{y_k}{2^k}} \ket{y} \\
    &\stackrel{(2)}{=}
    \frac{1}{\sqrt{N}} \sum_{y_1=0}^{1} \sum_{y_2=0}^{1} \dots \sum_{y_n=0}^{1} \left( \prod_{k=1}^{n} e^{i 2\pi x \times \frac{y_k}{2^k}} \ket{y_1 y_2 \dots y_n} \right) \\
    &=
    \frac{1}{\sqrt{N}} \sum_{y_1=0}^{1} \sum_{y_2=0}^{1} \dots \sum_{y_n=0}^{1} \left( e^{i 2\pi x \times \frac{y_1}{2^1}} e^{i 2\pi x \times \frac{y_2}{2^2}} \dots e^{i 2\pi x \times \frac{y_n}{2^n}} \ket{y_1 y_2 \dots y_n} \right) \\
    &=
    \frac{1}{\sqrt{N}} \sum_{y_1=0}^{1} \sum_{y_2=0}^{1} \dots \sum_{y_n=0}^{1} \left( 
                                                                                        e^{i 2\pi x \times \frac{y_1}{2^1}} \ket{y_1} \otimes 
                                                                                        e^{i 2\pi x \times \frac{y_2}{2^2}} \ket{y_2} \dots 
                                                                                        e^{i 2\pi x \times \frac{y_n}{2^n}} \ket{y_n} 
                                                                                \right) \\
    &=
    \frac{1}{\sqrt{N}} \sum_{y_1=0}^{1} e^{i 2\pi x \times \frac{y_1}{2^1}} \ket{y_1} \otimes 
                       \sum_{y_2=0}^{1} e^{i 2\pi x \times \frac{y_2}{2^2}} \ket{y_2} \otimes \dots \otimes 
                       \sum_{y_n=0}^{1} e^{i 2\pi x \times \frac{y_n}{2^n}} \ket{y_n} \\
    &=
    \frac{1}{\sqrt{N}} \left( \ket{0} + e^{i 2\pi \frac{x}{2^1}} \ket{1} \right) \otimes
                       \left( \ket{0} + e^{i 2\pi \frac{x}{2^2}} \ket{1} \right) \otimes \dots \otimes
                       \left( \ket{0} + e^{i 2\pi \frac{x}{2^n}} \ket{1} \right) \hspace*{2cm} (\star) \\
    &=
    \bigotimes_{k=1}^{n} \frac{1}{\sqrt{2}}  \left( \ket{0} + e^{i 2\pi \frac{x}{2^k}} \ket{1} \right)
\end{align*}

where in step $(1)$ we represent the integer $y$ as a sum of its binary terms, and in step $(2)$ we expand $y$ to its binary representation.

At this point, we observe that each basis state $\ket{1}$ acquires a phase factor $e^{i 2\pi \frac{x}{2^k}}$, while the state $\ket{0}$ remains unchanged.
This behavior is exactly that of a single-qubit phase gate acting only on $\ket{1}$, i.e. $$R_k = \begin{pmatrix} 1 & 0 \\ 0 & e^{i \frac{2\pi}{2^k}} \end{pmatrix}$$
Furthermore, in the phase $e^{i 2\pi \frac{x}{2^k}}$, only the last $k$ bits of the binary representation of $x$ contribute nontrivially.  The higher-order bits correspond to integer multiples of $2 \pi$ and therefore have no effect on the phase.

To build the circuit, it seems that we need some Hadamard gates and some rotation gates.
We want the rotation gates to act on the qubits in a controlled way.
This is better understood if we rewrite Equation $(\star)$ using the fractional binary notation:
\begin{align*}
    &
    \frac{1}{\sqrt{N}} \left( \ket{0} + e^{i 2\pi \frac{x}{2^1}} \ket{1} \right) \otimes
                       \left( \ket{0} + e^{i 2\pi \frac{x}{2^2}} \ket{1} \right) \otimes \dots \otimes
                       \left( \ket{0} + e^{i 2\pi \frac{x}{2^n}} \ket{1} \right) \\
    =&
    \frac{1}{\sqrt{N}} \left( \ket{0} + e^{i 2\pi 0.x_{n}} \ket{1} \right) \otimes
                       \left( \ket{0} + e^{i 2\pi 0.x_{n-1} x_{n}} \ket{1} \right) \otimes \dots \otimes
                       \left( \ket{0} + e^{i 2\pi 0. x_1 x_2 \dots x_n} \ket{1} \right)
\end{align*}

Note that increasing the number of qubits increases the precision with which phases are represented in the QFT. 
Each additional qubit allows the circuit to resolve finer fractional binary digits of a phase.
Accordingly, the controlled rotation gates are applied in a way that reflects this precision hierarchy. 
For the $k$-th output qubit, the accumulated phase depends on the values of the less significant input qubits. 
In particular, for the $n$-th qubit, whose phase is the most sensitive, we must apply a sequence of controlled rotations conditioned on the states of the previous $n-1$ qubits. 
Each of these controlled rotations contributes a progressively smaller phase, corresponding to increasingly fine binary fractions.
In sum: each qubit refines the phase estimate by one binary digit, and the controlled rotations ensure that this refinement is applied in the correct order.

## PennyLane implementation of the QTF

At this point, we drop theory a little bit and focus on the implementation of the QTF in PennyLane.

```python
import pennylane as qml
from pennylane import numpy as np
import math
```

```python
def state_to_ket(state, *, tol=1e-10, max_terms=None, binary=True, normalize_check=False):
    """
    Convert a statevector to Dirac ket notation.

    Parameters
    ----------
    state : array-like of complex
        1D array of amplitudes, length = 2^n (typically).
    tol : float
        Amplitudes with |amp| <= tol are omitted.
    max_terms : int or None
        If set, shows only the largest `max_terms` amplitudes by magnitude.
    binary : bool
        If True, basis labels are |00...1>. If False, labels are |k>.
    normalize_check : bool
        If True, raises ValueError if state is not ~normalized.

    Returns
    -------
    str : a Dirac-notation string
    """
    v = np.asarray(state, dtype=complex).flatten()
    if v.ndim != 1:
        raise ValueError("state must be a 1D array-like.")
    dim = v.size
    if dim == 0:
        return "0"

    # Optional normalization check
    if normalize_check:
        norm = float(np.vdot(v, v).real)
        if not isclose(norm, 1.0, rel_tol=1e-9, abs_tol=1e-9):
            raise ValueError(f"State not normalized: ||ψ||^2 = {norm}")

    # Determine qubit count if possible
    n = None
    if dim & (dim - 1) == 0:  # power of two
        n = int(log2(dim))

    # Collect nonzero terms
    terms = [(i, amp) for i, amp in enumerate(v) if abs(amp) > tol]
    if not terms:
        return "0"

    # Optionally keep only largest amplitudes
    if max_terms is not None and max_terms > 0 and len(terms) > max_terms:
        terms.sort(key=lambda x: abs(x[1]), reverse=True)
        terms = terms[:max_terms]
        terms.sort(key=lambda x: x[0])

    def fmt_complex(z):
        # Nicely format complex numbers with small real/imag suppressed.
        r = 0.0 if abs(z.real) < tol else z.real
        im = 0.0 if abs(z.imag) < tol else z.imag
        if im == 0.0:
            return f"{r:.6g}"
        if r == 0.0:
            return f"{im:.6g}j"
        sign = "+" if im >= 0 else "-"
        return f"({r:.6g}{sign}{abs(im):.6g}j)"

    def basis_label(i):
        if binary:
            if n is None:
                return f"|{i}⟩"
            return f"|{format(i, f'0{n}b')}⟩"
        return f"|{i}⟩"

    pieces = []
    for i, amp in terms:
        label = basis_label(i)

        # Handle special cases for amplitude formatting
        if abs(amp - 1) <= tol:
            coeff = ""
            sign = "+"
        elif abs(amp + 1) <= tol:
            coeff = ""
            sign = "-"
        else:
            coeff = fmt_complex(amp)
            sign = "+"

        pieces.append((sign, coeff, label))

    # Build string with proper leading sign
    out = []
    for idx, (sign, coeff, label) in enumerate(pieces):
        if idx == 0:
            if sign == "-":
                out.append("-")
            # if coeff empty, just label; else coeff + label
            out.append(f"{coeff}{label}" if coeff else f"{label}")
        else:
            out.append(f" {sign} ")
            out.append(f"{coeff}{label}" if coeff else f"{label}")

    return "".join(out)


def print_state_ket(state, **kwargs):
    """Convenience: print the ket string."""
    print(state_to_ket(state, **kwargs))
```

We shall implement the QFT in three different ways:
1. matrix
2. iterative
3. recursive

Starting with the *matrix method*. 
We use the function *qml.BasisState* to prepare the input state for the QFT

```python
num_wires = 2
dev = qml.device("default.qubit", wires=num_wires)

@qml.qnode(dev)
def matrix_QFT(basis_id):
    # Prepare the basis state |basis_id>
    bits = [int(x) for x in np.binary_repr(basis_id, width=num_wires)]
    qml.BasisState(bits, wires=range(num_wires))

    matrix = []
    N=2**num_wires
    const= (1/np.sqrt(N)).item()
    #Building the matrix of the QFT
    for i in range(N):
        row=[]
        for j in range(N):
            row.append(const*(1j)**(i*j))
        matrix.append(row)

    #Printing the matrix of the QFT
    for i in range(N):
        for j in range(N):
            print(matrix[i][j], end=" ")
        print()

    #Applying the QFT's matrix to the initial state
    qml.QubitUnitary(matrix, wires=range(num_wires))
            

    return qml.state()

matrix_QFT(3)
```

We now go to the iterative method

```python
num_wires = 4
dev_qft = qml.device("default.qubit", wires=num_wires)
```

```python
@qml.qnode(dev_qft)
def it_qft(n_qubits):
    for q in range(n_qubits):
        qml.H(q)
        for i in range(q+1,n_qubits):
            qml.ControlledPhaseShift((2*np.pi)/(2**(2+(i-(q+1)))),wires=[i,q])
    for i in range(n_qubits//2):
        qml.SWAP(wires=[i,n_qubits-1-i])

    return qml.state(), qml.probs()
print(qml.draw(it_qft)(3))
```

The recursive method is as follows

```python
def qft_recursive_rotations(n_qubits, wire=0):
    """A circuit that performs the QFT rotations on the specified qubits
        recursively.
        
    Args:
        n_qubits (int): An integer value identifying the number of qubits.
        wire (int): An integer identifying the wire 
                    (or the qubit) to apply rotations on.
    """
    if (wire==(n_qubits)-1):
        qml.H(wire)
    else:
        qml.H(wire)
        for i in range(wire+1, n_qubits):
            qml.ControlledPhaseShift((2*np.pi)/(2**(2+(i-(wire+1)))),wires=[i,wire])
        qft_recursive_rotations(n_qubits,wire=wire+1)

#@qml.qnode(dev_qft)
def rec_qft(n_qubits):
    qft_recursive_rotations(n_qubits,0)
    for i in range(n_qubits//2):
        qml.SWAP(wires=[i,n_qubits-1-i])

    #return qml.state()

#print(qml.draw(qft_recursive_rotations)(4,0))
print(qml.draw(rec_qft)(4))


```

## Period Finding

Let us now use the QFT to find the period of a periodic function.
The problem is as follows:
* Given: 
    * A black box that implements a function $f: \mathbb{B}^{n} \rightarrow \mathbb{B}^{m}$
* Assumptions: 
    * The function $f$ is periodic with period $r \in \{2, \dots, 2^{n-1}\}$
    * The function assigns different outputs to elements of the domain within a period
* Goal:
    * Estimate the period of $f$, i.e. the smallest integer $r$ satisfying $f(x) = f(x + r)$

Here we are going to implement the algorithm, and we let the theoretical foundations for later.

```python
dev_qft_period = qml.device("default.qubit",wires=3, shots=10)

#For each function f we are considering, we need to provide the respective codification into a unitary u_f
# f(00) = f(10) = 0, f(01) = f(11) = 1
# u_f|x>|y> = |x>|y (+) f(x)>  

f_test = [("00","0"),("01","1"),("10","0"),("11","1")]

def to_matrix(f_pairs, n, m, dtype=np.complex128):
    """
    Build the oracle unitary U_f for a function f encoded as list of pairs
    [(x_bits, fx_bits), ...], where x_bits and fx_bits are binary strings.

    Basis ordering: |x,y> with index = (x << m) | y  (y are the low bits).
    Oracle action: U_f |x>|y> = |x>|y XOR f(x)>

    Returns a 2^(n+m) x 2^(n+m) permutation matrix (numpy array).
    """
    if not f_pairs:
        raise ValueError("f_pairs must be non-empty")

    # build mapping x_int -> fx_int
    mapping = {}
    for x_bits, fx_bits in f_pairs:
        if len(x_bits) != n or len(fx_bits) != m:
            raise ValueError("Inconsistent bit-lengths in f_pairs")
        if any(c not in "01" for c in x_bits + fx_bits):
            raise ValueError("Inputs must be binary strings")

        x_int = int(x_bits, 2)
        fx_int = int(fx_bits, 2)
        if x_int in mapping and mapping[x_int] != fx_int:
            raise ValueError(f"Conflicting values for input {x_bits}")
        mapping[x_int] = fx_int

    N = 1 << n #2**n
    M = 1 << m #2**m

    # ensure f defined on all x
    missing = [x for x in range(N) if x not in mapping]
    if missing:
        raise ValueError("f not defined for inputs: " +
                         ", ".join(format(x, f"0{n}b") for x in missing))

    dim = N * M
    U = np.zeros((dim, dim), dtype=dtype)

    for x in range(N):
        fx = mapping[x]
        for y in range(M):
            in_idx  = (x << m) | y         # = x * 2^m + y
            out_idx = (x << m) | (y ^ fx)  # y XOR f(x) in the low bits
            U[out_idx, in_idx] = 1.0

    return U


def continued_fraction_expansion(x):
    #x is a real number
    #limit is the number of coefficients used to round the real number x with the continued fraction expansion
    limit = 0
    coeff = []
    a1 = math.floor(x)
    x1 = x
    coeff.append(a1)
    while(a1-x1!=0):
        if(limit==10) : break
        x2 = 1/(x1 - a1)
        a1 = math.floor(x2)
        coeff.append(a1)
        x1=x2
        limit=limit+1
    return coeff

def convergents(coeff):
    #coeff is a list of coefficients obtained from the continued fraction expansion
    conv = []
    n=len(coeff)
    p1,p2 = 1,0
    q1,q2 = 0,1

    for a in coeff:
        p = a*p1 + p2
        q = a*q1 + q2 
        conv.append((p,q))
        p1,p2 = p, p1
        q1,q2 = q, q1
    
    return conv


def check_convergents_condition(conv,x,n):
    #conv is the list of convergents
    #x is a value from a measurement
    #n is the number of qubits

    bool_list = []
    N = 2**n
    for (p,q) in conv:
        b1 = abs((x/N) - (p/q)) <= 1/(2 * (q**2))
        b2 = 1 < q**2 < N
        print(p, q, b1, b2)
        bool_list.append(b1 and b2)
    return bool_list

@qml.qnode(dev_qft_period)
def period_finding(f, n_qubits_in, n_qubits_out):
    # f is a function that has n qubits as input, n_qubits_in, and m qubits as output, n_qubits_out
    # n_qubits_in and n_qubits_out are integers

    n_total = n_qubits_in + n_qubits_out

    #apply hadamard to n_qubits_in
    for i in range(n_qubits_in):
        qml.H(i)

    #apply the unitary function u_f that encodes the function f
    u_f = to_matrix(f,n_qubits_in,n_qubits_out)
    qml.QubitUnitary(u_f,wires=range(n_total))
    
    rec_qft(n_qubits_in)

    #return qml.probs(wires=range(n_qubits_in))
    return qml.sample(wires=range(n_qubits_in))

def bits_to_int(bits):
    out = 0
    n=len(bits)-1
    for b in bits:
        out= out + int(b)*2**n
        n=n-1
    #for b in bits:
    #    out = (out << 1) | int(b)
    return out


def period_from_samples(samples, n):
    N = 1 << n
    candidates = []

    for s in samples:
        # If your QFT is bit-reversed, use s[::-1]
        k = bits_to_int(s)   # or bits_to_int(s[::-1])
        if k == 0:
            continue
        coeffs = continued_fraction_expansion(k/N)
        convs = convergents(coeffs)    # your convergents() works fine here
        candidates.append(convs[-1][1])  # denominator of best convergent

    return candidates


#print(qml.draw(period_finding)(f_test,2,1))
#print(period_finding(f_test,2,1))

convs = convergents(continued_fraction_expansion(2/4))
print(convs)

print(check_convergents_condition(convs,2,2))

#print(bits_to_int([1,1]))

#samples = period_finding(f_test,2,1)
#period_from_samples(samples, 2)
```

### Relation between Dirac notation and matrices

Let $U$ be a unitary operator action on an $n$-dimensional Hilbert space with computational basis $\{\ket{0}, \ket{1}, \dots, \ket{n-1}\}$.

The matrix representation of $U$ in this basis is the $n \times n$ matrix whose entries are given by $$U[i][j] = \bra{i} U \ket{j}$$
where $i.j \in \{0, \dots, n-1\}$.

This implies the following correspondence:
* the column index $j$ labels the input basis state $\ket{j}$
* the row index $i$ labels the output basis state $\ket{i}$

Explicitly, the matrix of $U$ can be written as
\begin{align*}
    U = 
    \begin{pmatrix}
        u_{0,0} & u_{0,1} & \dots & u_{0,n-1} \\
        u_{1,0} & u_{1,1} & \dots & u_{1,n-1} \\
        \vdots & \vdots & \ddots & \vdots \\
        u_{n-1,0} & u_{n-1,1} & \dots & u_{n-1,n-1} \\
    \end{pmatrix}
    =
    \begin{pmatrix}
        \bra{0} U \ket{0} & \bra{0} U \ket{1} & \dots & \bra{0} U \ket{n-1} \\
        \bra{1} U \ket{0} & \bra{1} U \ket{1} & \dots & \bra{1} U \ket{n-1} \\
        \vdots & \vdots & \ddots & \vdots \\
        \bra{n-1} U \ket{0} & \bra{n-1} U \ket{1} & \dots & \bra{n-1} U \ket{n-1} \\
    \end{pmatrix}
\end{align*}

Equivalently, the operator $U$ can be reconstructed from its matrix elements as $$U = \sum_{i=0}^{n-1} \sum_{j=0}^{n-1} \bra{i} U \ket{j} \ket{i}\bra{j}$$


Furthermore, as a mental note, the sum $\sum_{i=0}^{n-1} \sum_{j=0}^{n-1} \ket{i}\bra{j}$ gives the following matrix:
\begin{align*}
    \begin{pmatrix}
        \ket{0}\bra{0} & \ket{0}\bra{1} & \dots & \ket{0}\bra{n-1} \\
        \ket{1}\bra{0} & \ket{1}\bra{1} & \dots & \ket{1}\bra{n-1} \\
        \vdots & \vdots & \ddots & \vdots \\
        \ket{n-1}\bra{0} & \ket{n-1}\bra{1} & \dots & \ket{n-1}\bra{n-1} \\
    \end{pmatrix}
\end{align*}


To map a function $$f : \{0,1\}^{n} \rightarrow \{0,1\}^{m}$$ to a unitary oracle matrix $U_f$, the following points must be taken into account:
* the dimension of the matrix $U_f$ is $2^{n+m}$
* the action of the oracle is $U_f \ket{x}\ket{y} = \ket{x}\ket{y \oplus f(x)}$
* from $U\ket{input} = \ket{output}$, we can say the following:
    * elements $\ket{input}$ are extended by a factor of $\ket{input}*2^m + y$, where $y \in \{0,1\}^{m}$
    * elements $\ket{output}$ are extended by a factor of $\ket{input}*2^m + (y \oplus f(\ket{input}))$

```python
f2 = [("0","1"),("1","0")]

def to_matrix_aux(f,n,m,dtype=np.complex128):
    N = 2**n 
    M = 2**m 
    dim = 2**(n+m)
    U = np.zeros((dim,dim),dtype=dtype)
    for x in range(N):
        fx = f[x][1]
        for y in range(M):
            id_in = x*M + y 
            id_out = x*M + (y^int(fx,2)) 
            U[id_in][id_out] = 1.0
    return U

print(to_matrix(f2,1,1))
print(to_matrix_aux(f2,1,1))
print(to_matrix(f2,1,1) == to_matrix_aux(f2,1,1))
```

```python
def my_fft(poly_a,poly_b):
    d_a = len(poly_a)-1 #degree of poly_a
    d_b = len(poly_b)-1 #degree of poly_b
    d_c = d_a+d_b #degree of poly_c
    v_c = d_c+1 #points needed for poly_c with degree d_c

    np2 = 2**int(np.ceil(np.log2(v_c))) #points for poly_c based on power of 2

    polyA = np.pad(poly_a, (0,np2-(d_a+1)), 'constant')
    polyB = np.pad(poly_b, (0,np2-(d_b+1)), 'constant')

    vrA = np.fft.fft(polyA, len(polyA))
    vrB = np.fft.fft(polyB, len(polyB))

    vrC = vrA*vrB

    polyC = np.fft.ifft(vrC)

    return polyC
```

```python
print(my_fft([1,1],[1,-1]))
```

```python
num_wires = 2
dev = qml.device("default.qubit", wires=num_wires)

@qml.transforms.undo_swaps
@qml.qnode(dev)
def decompose_two_qubit_QFT(basis_id):
    """A circuit that computes the QFT on two qubits using elementary gates.
    
    Args:
        basis_id (int): An integer value identifying the basis state to construct.
    
    Returns:
        array[complex]: The state of the qubits after the QFT operation.
    """
    # Prepare the basis state |basis_id>
    bits = [int(x) for x in np.binary_repr(basis_id, width=num_wires)]
    qml.BasisState(bits, wires=[0, 1])
    
    qml.H(0)
    qml.ctrl(qml.S,1,1)(0)
    qml.H(1)
    #qml.SWAP([0,1])
    
    # qml.H(1)
    # qml.ctrl(qml.S,0,1)(1)
    # qml.H(0)
    
    return qml.state()

decompose_two_qubit_QFT(2)
```
